```python
#import
import mlflow
from ultralytics import YOLO
```

```python
#MLflow Remote Setup while tracking from local
mlflow.set_tracking_uri("https://take-your-ngrok.ngrok-free.dev")
mlflow.set_experiment("YOLO-Colab-Remote-Tracking")
```

```python
#mount drive to access the dataset
from google.colab import drive
drive.mount('/content/drive')
```

#========================================
#YOLOv8 Nano
#========================================

```python

#load the model (yolo8n)
model_file = "yolov8n.pt"
#set hypherparameter
params = {
    "epochs": 5,
    "batch": 16,
    "lr0": 0.01
}

# Start MLflow run for this model
with mlflow.start_run(run_name="YOLOv8n"):
    print(f"Training {model_file}")

    # Log parameters
    mlflow.log_param("model", model_file)
    for k, v in params.items():
        mlflow.log_param(k, v)

    model = YOLO(model_file)

     # Initialize and train
    results = model.train(
        data="/content/drive/MyDrive/SESAME Plant Detection Dataset/data.yaml",
        epochs=params["epochs"],
        batch=params["batch"],
        lr0=params["lr0"]
    )

     # Log metrics
    mlflow.log_metric("mAP50", results.box.map50)
    mlflow.log_metric("mAP50-95", results.box.map)
    mlflow.log_metric("precision", results.box.mp)
    mlflow.log_metric("recall", results.box.mr)

    #log model
    model.save("yolov8_model.pt")
    mlflow.log_artifact("yolov8_model.pt")

print("YOLOv8n logged successfully")
```

#========================================
#YOLO11 Nano
#========================================

```python
#log the model (yolo11n)
model_file = "yolo11n.pt"

#set hyperparameter
params = {
    "epochs": 5,
    "batch": 12,
    "lr0": 0.005
}

#initialize and train
with mlflow.start_run(run_name="YOLO11n", nested=True):
    print(f"Training {model_file}")

     #log hyperparameter
    mlflow.log_param("model", model_file)
    for k, v in params.items():
        mlflow.log_param(k, v)

    model = YOLO(model_file)

    #initialize and train
    results = model.train(
        data="/content/drive/MyDrive/SESAME Plant Detection Dataset/data.yaml",
        device=0,
        epochs=params["epochs"],
        batch=params["batch"],
        lr0=params["lr0"]
    )

    #log metrics
    mlflow.log_metric("mAP50", results.box.map50)
    mlflow.log_metric("mAP50-95", results.box.map)
    mlflow.log_metric("precision", results.box.mp)
    mlflow.log_metric("recall", results.box.mr)

    #log m
    model.save("yolov8_model.pt")
    mlflow.log_artifact("yolov8_model.pt")
print("YOLO11n logged successfully")
```
